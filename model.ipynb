# Import Module

```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
```

# Load Data
## Declare filename

```python
# Define a list of filenames for each object (replace with actual filenames)
pencil_sharpener_files = ["Pencil_Sharpener_Augmented_"+str(i+1)+".csv" for i in range(5)]
nothing_files = ["Nothing_Augmented_"+str(i+1)+".csv" for i in range(5)]
clip_files = ["Clip_Augmented_"+str(i+1)+".csv" for i in range(5)]
staple_files = ["Staple_Augmented_"+str(i+1)+".csv" for i in range(5)]
```

```python
pencil_sharpener_files
```

```python
Clip =[]
Nothing = []
Pencil = []
Staple = []
```

## Store data into those list

```python
for i in range(5): 
        X__clip = np.row_stack([
                # By indexing with [None], you're using numpy's None indexing feature (also known as np.newaxis), which adds a new axis to an array
                group.drop(columns=['Augmentation','Object','window_label']).values[None] # drop Augmentation, Object, window_label after grouping
                for _, group in pd.read_csv(clip_files[i]).groupby('window_label')]) # group name, grouped dataframe
        Clip.append(X__clip)

        X__pencil_sharpener = np.row_stack([
                # By indexing with [None], you're using numpy's None indexing feature (also known as np.newaxis), which adds a new axis to an array
                group.drop(columns=['Augmentation','Object','window_label']).values[None] # drop series_id & measurement_number after grouping
                for _, group in pd.read_csv(pencil_sharpener_files[i]).groupby('window_label')]) # group name, grouped dataframe
        Pencil.append(X__pencil_sharpener)

        X___nothing = np.row_stack([
                # By indexing with [None], you're using numpy's None indexing feature (also known as np.newaxis), which adds a new axis to an array
                group.drop(columns=['Augmentation','Object','window_label']).values[None] # drop series_id & measurement_number after grouping
                for _, group in pd.read_csv(nothing_files[i]).groupby('window_label')]) # group name, grouped dataframe
        Nothing.append(X___nothing)

        X___staple = np.row_stack([
                # By indexing with [None], you're using numpy's None indexing feature (also known as np.newaxis), which adds a new axis to an array
                group.drop(columns=['Augmentation','Object','window_label']).values[None] # drop series_id & measurement_number after grouping
                for _, group in pd.read_csv(staple_files[i]).groupby('window_label')]) # group name, grouped dataframe
        Staple.append(X___staple)
```

```python
Clip = np.vstack(Clip)
Clip
```

```python
X = np.concatenate([
    X__clip, 
    X__pencil_sharpener, 
    X___nothing, 
    X___staple
], axis=0)

y = np.concatenate([
    np.full(X__clip.shape[0], 0),  # Clip
    np.full(X__pencil_sharpener.shape[0], 1),  # Pencil Sharpener
    np.full(X___nothing.shape[0], 2),  # Nothing
    np.full(X___staple.shape[0], 3)   # Staple
])
```
